```python
# Importing Packages
import time
import random
import numpy as np
import pandas as pd
```

```python
# Generating values for n
n = []
for x in range(5):
    n.append(10*(x+1))
print(n)
# Generating value for k
k = 5
print(k)
```

```python
# Creating a function called min_cost_with_limited_astro_haunted_galaxies
def min_cost_with_limited_astro_haunted_galaxies(astro_haunted,galaxy_distances,n,k):
    # Initialize the D matrix for memoization
    D = [[[float('inf') for _ in range(n)] for _ in range(n)] for _ in range(k + 1)]
    #Base Case
    for i in range(n):
        for j in range(n):
            if i != j and astro_haunted[i] != 1 and astro_haunted[j] != 1:
                D[0][i][j] = galaxy_distances[i][j]
    # Recursive call
    for h in range(1, k + 1):
        for i in range(n):
            for j in range(n):
                D[h][i][j] = D[h - 1][i][j]
                for z in range(n):
                    if astro_haunted[z] == 1:
                        D[h][i][j] = min(D[h][i][j], D[h - 1][i][z] + D[0][z][j])
    # Return the shortest distance from 1 to n with at most k astro-haunted galaxies
    return D[k][0][n - 1]
```

```python
output = []
time_taken = []
for idx, y in enumerate(n):
    # Creating a matrix with the distaces between the galaxies
    galaxy_distances = np.random.randint(1, 10, size=(y, y))
    # Ensuring that the distance to itself is 0 by initializaing the diagonal on te matrix to 0
    np.fill_diagonal(galaxy_distances, 0)
    # Creating a list of astro-haunted galaxies
    astro_haunted = np.random.randint(0, 2, size=y)
    # Converting the first and last galaxy into non-astro-haunted galaxies
    astro_haunted[0] = 0
    astro_haunted[-1] = 0
    # Start time for the function
    start_time = time.time()
    # Run the min_cost_with_limited_astro_haunted_galaxies function and append it
    output.append(min_cost_with_limited_astro_haunted_galaxies(astro_haunted,galaxy_distances,y,k))
    # End time for the function
    end_time = time.time()
    # Creating a list for the time taken
    time_taken.append(end_time-start_time)
```

```python
# Creating a dataframe with the data
dictionary = {"n":n, "k":k,"min_cost_with_limited_astro_haunted_galaxies":output, "Time_taken_in_seconds": time_taken}
df = pd.DataFrame(dictionary,columns=["n","k","min_cost_with_limited_astro_haunted_galaxies","Time_taken_in_seconds"])
df
```
